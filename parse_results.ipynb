```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mainpath = "/home/ubuntu/simulations_my_svd/results"
```

```python
dime = 64

accs = []


for i in range(30,33):
    accs.append({})
    for part in ['train', 'valid', 'ttest']:
        accs[-1][part] = np.load(f"{mainpath}/accs_{part}_CiteSeer_realization_{i}_embed_dime_{dime}_dimy_16.npy")
```

```python
total = 0
for i in range(len(accs)):
    val = accs[i]['ttest']
    print(val)
    total += val
total /= len(accs)
print()
print(total)
```
